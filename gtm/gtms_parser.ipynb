```python
from bs4 import BeautifulSoup as bs
import os
import pandas as pd
```

```python
filepath = os.path.join("gtms.html")
with open(filepath) as file:
    html = file.read()
```

```python
soup = bs(html, 'lxml')
```

```python
td = soup.find_all('td')
print(td)
```

```python
a = soup.find_all('a')
```

```python
links_with_text = []
name = []

for a in soup.find_all('a', href=True): 
    if a.text: 
        links_with_text.append(a['href'])
        name.append(a.text)
```

```python
d = {'Name': name, 'link': links_with_text}
```

```python
df = pd.DataFrame(d)
# df.to_csv('gtm_list.csv')
```
